```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Input widgets
name_input = widgets.Text(
    value='',
    placeholder='Enter student name',
    description='Name:',
    disabled=False
)

mark_inputs = [
    widgets.BoundedFloatText(
        value=0.0,
        min=0.0,
        max=100.0,
        step=0.1,
        description=f'Subject {i+1}:',
        disabled=False
    ) for i in range(5)
]

calculate_button = widgets.Button(
    description='Calculate Grade',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to calculate grade',
    icon='check'
)

output_widget = widgets.Output()

# Grade calculation function
def calculate_grade(name, marks):
    total = sum(marks)
    percentage = total / len(marks)

    if percentage >= 90:
        grade = "A"
    elif percentage >= 75:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    elif percentage >= 45:
        grade = "D"
    else:
        grade = "F"

    return (
        f"Name: {name}\n"
        f"Total Marks: {total}\n"
        f"Percentage: {percentage:.2f}%\n"
        f"Grade: {grade}"
    )

# Event handler for the button click
def on_calculate_button_clicked(b):
    with output_widget:
        clear_output()
        student_name = name_input.value
        student_marks = [m.value for m in mark_inputs]
        if student_name and len(student_marks) == 5:
            result = calculate_grade(student_name, student_marks)
            print(result)
        else:
            print("Please enter student name and all 5 marks.")

# Attach the event handler to the button
calculate_button.on_click(on_calculate_button_clicked)

# Display the widgets
display(name_input)
print("Enter marks for 5 subjects (0-100):")
for mark_input in mark_inputs:
    display(mark_input)
display(calculate_button)
display(output_widget)
```
